```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importar os dados do DataSet e Visualizar
```

```python
pf = pd.read_csv("Data.csv");
pf
```

```python
#DataSet com 4 colunas
#Country - variável categórica
#Age - variável numérica com dados faltantes
#Salary - variável numérica com dados faltantes
#Purchased - variável-resposta categórica
```

```python
#visualização dtypes
pf.dtypes
```

```python
#histograma
fig, ax = plt.subplots(ncols=2, figsize=(12,5))
pf.hist(column="Age", ax=ax[0], bins=5)
pf.hist(column="Salary", ax=ax[1], bins=5)
plt.show()
```

```python
#boxplot
fig, ax = plt.subplots(ncols=2, figsize=(12,5))
pf.boxplot(column="Age", ax=ax[0])
ax[0].set_ylabel("Idade")
pf.boxplot(column="Salary", ax=ax[1])
ax[1].set_ylabel("Salário (US$)")
plt.show()
```

```python
#scatter plot - relação linear entre variávies
pf.plot.scatter(x="Age", y="Salary", grid="on");
```

```python
#seaborn para visualização automática
sns.pairplot(data=pf[['Age','Salary']], diag_kind="hist");
```

```python
#seaborn para visualização automática
sns.pairplot(data=pf[['Age','Salary']], diag_kind="kde");
```

```python
#seaborn para visualização automática
sns.pairplot(data=pf[['Age','Salary']], kind="reg");
```

```python
#seaborn para visualização automática
sns.pairplot(data=pf[['Age','Salary']], kind="reg", diag_kind="kde");
```
